**Part below is for looping through all files**

```python
google_dir = ('/content/drive/MyDrive/CNNRelatedFolders/FinalDatasetConverted/pork_chop/')

data_dir = pathlib.Path('/content/drive/MyDrive/CNNRelatedFolders/FinalDataset/pork_chop').glob('**/*')
data_dir

for path in data_dir:
  # get filename and use as the image name
  filepath = pathlib.Path(path).name #get complete name of the file (ex: img_name.NEF)
  filename, file_extension = os.path.splitext(filepath) #split the name and remove the file extension

  final_filename = google_dir + filename + '.jpeg' #create the whole path where the image will be saved; must be string
  print(final_filename)

  with rawpy.imread(str(path)) as raw:
    # raises rawpy.LibRawNoThumbnailError if thumbnail missing
    # raises rawpy.LibRawUnsupportedThumbnailError if unsupported format
    thumb = raw.extract_thumb()
  if thumb.format == rawpy.ThumbFormat.JPEG:
      # thumb.data is already in JPEG format, save as-is
      with open(final_filename, 'wb') as f:
          f.write(thumb.data)
  elif thumb.format == rawpy.ThumbFormat.BITMAP:
      # thumb.data is an RGB numpy array, convert with imageio
      imageio.imsave(final_filename, thumb.data)
  
```

# ***Code below is for converting dng file format***

```python
google_dir = ('/content/drive/MyDrive/CNNRelatedFolders/TestDataset/chicken_breast/')

directory = '/content/drive/MyDrive/CNNRelatedFolders/FinalDataset/dng_files/chicken_breast/'

for filename in os.listdir(directory):
  # get filename and use as the image name
  file_directory = os.path.join(directory, filename)
  split_filename, file_extension = os.path.splitext(filename)
  final_filename = google_dir + split_filename + '.jpeg'
  with rawpy.imread(file_directory) as raw:
    rgb = raw.postprocess()
  imageio.imsave(final_filename, rgb)
  print(final_filename)
  
  
```
